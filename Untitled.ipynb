```python
def text_analysis(paragraph):
    paragraph=paragraph.strip()

    sentences=0
    for ch in paragraph:
        if ch in ".!?":
            sentences+=1
    words=paragraph.replace(',','').replace('?','').replace('!','').split()
    freq={}
    for word in words:
        word_lower=word.lower()
        freq[word_lower]=freq.get(word_lower,0)+1
    longest_word=max(words, key=len)
    print("\n-------Text Analysis----")
    print(f"Number of sentences is: {sentences}")
    print(f"Number of words is: {len(words)}")
    print(f"Longest word is {longest_word}")
    print("\n Word Frequencies ")
    for word,count in freq.items():
        print(f"{word}:{count}")
def main():
    paragraph=input("enter a paragraph")
    text_analysis(paragraph)
if __name__=="__main__":
    main()
    
```

```python
class complex_number:
    def __init__ (self,real,imag):
        self.real=real
        self.imag=imag
    def add(self,other):
        return complex_number(self.real+other.real,self.imag+other.imag)
    def __str__ (self):
        sign ="+" if self.imag >=0 else "-"
        return f"{self.real} {sign} {abs(self.imag)}i"
def main():
    n=int(input("\n Enter the number of complex numbers to be added."))
    total=complex_number(0,0)
    for i in range (n):
        r=float(input("enter real part of complex number"))
        im=float(input("enter imaginary part of complex number"))
        total=total.add(complex_number(r,im))
    print(f"The sum of {n} complex number is {total}")
if __name__=="__main__":
    main()
```

```python
class employee:
    def __init__ (self,name,salary):
        self.name=name
        self.salary=salary
    def display(self):
        print(f"Employee Name:{self.name},Employee Salary:{self.salary}")
class clerk(employee):#derived class 1. 
    def __init__(self,name,salary,department):
        super().__init__(name,salary)
        self.department=department
    def display(self):
        print(f"Clerk name:{self.name}, clerk salary:{self.salary}, clerk department:{self.department}")
class SoftwareEngineer(employee):
    def __init__(self,name,salary,language):
        super().__init__(name,salary)
        self.language=language
    def display(self):
        print(f"The software engineer name: {self.name}, software engineer salary: {self.salary}, software engineer language is {self.language}")
class Teamleader(SoftwareEngineer):
    def __init__ (self,name,salary,language,team_size):
        super().__init__(name,salary,language)
        self.team_size=team_size
    def display(self):
        print(f"The team leader name is {self.name}, team leader salary is {self.salary}, team leader language is {self.language}, team leader team size is {self.team_size}")
def main():
    c=clerk("John",13000,"CSE")
    se=SoftwareEngineer("John",12039,"Python")
    Te=Teamleader("S",12345,"Jav",123)
    c.display()
    se.display()
    Te.display()
if __name__=="__main__":
    main()
    
```
